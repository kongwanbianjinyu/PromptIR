```python
import matplotlib.pyplot as plt
import numpy as np


# Plotting
fig, ax = plt.subplots(figsize=(12,10))
bar_width = 0.35
index = np.arange(len(months))

bars1 = ax.bar(index, fake_fee, bar_width, label='Fake Fee')
bars2 = ax.bar(index + bar_width, real_fee, bar_width, label='Real Fee')

# Setting x-axis labels
months = ["2023/09", "2023/10", "2023/11", "2024/01", "2024/02", "2024/03", "2024/04"]
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(months)

# Adding numbers on top of the bars
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{}'.format(height),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

ax.set_xlabel('Month')
ax.set_ylabel('Money($)')
ax.set_title('Fee Comparison')
ax.legend()


plt.savefig('fee_comparison.png', dpi=300, bbox_inches='tight')

plt.show()
```
